```python
def excel_to_dict_array(file_path, sheet_name = 0):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Excel file not found at: {file_path}")
    
    try:
        df = pd.read_excel(
            file_path,
            sheet_name=sheet_name,
            na_filter=True,
            keep_default_na=True
        )
        if df.empty:
            raise ValueError("Excel file is empty")

        df = df.where(pd.notna(df), None)
        dict_array = df.to_dict(orient='records')
        
        return dict_array

    except ValueError as ve:
        raise ValueError(f"Error reading Excel file: {str(ve)}")
    except Exception as e:
        raise Exception(f"Unexpected error while processing Excel file: {str(e)}")

def analyze_frv_decrease(data, x):
    sorted_data = sorted(data, key=lambda d: d['ano'], reverse=True)
    
    if x >= len(sorted_data):
        raise ValueError(f"Requested {x} years of analysis but only have {len(sorted_data)} years of data")
    
    end_year_frv = sorted_data[0]['frv']
    start_year_frv = sorted_data[x]['frv']
    
    total_percent_change = (end_year_frv - start_year_frv) / start_year_frv
    
    annual_rate = (pow(end_year_frv / start_year_frv, 1/x) - 1)
    
    return total_percent_change*100, annual_rate*100


def get_municipality_names():
    all_data = []
    base_url = 'http://www.ssp.sp.gov.br/v1/Municipios/RecuperaMunicipios'    
    response = requests.get(base_url)
    response.raise_for_status()
    data = response.json()

    data = data.get("data")
    city_names = []
    for city in data:
        city_names += [city.get("nome")]
    
    return city_names

async def fetch_municipality(session: aiohttp.ClientSession, municipality_id: int) -> List[Dict]:
    base_url = 'http://www.ssp.sp.gov.br/v1/OcorrenciasAnuais/recuperaDadosMunicipio'
    params = {'idMunicipio': municipality_id}
    
    try:
        async with session.get(base_url, params=params) as response:
            response.raise_for_status()
            data = await response.json()
            
            municipality_data = data.get('data', [])
            if municipality_data:
                logger.debug(f"Successfully fetched data for municipality {municipality_id}")
                return municipality_data
            else:
                logger.debug(f"No data found for municipality {municipality_id}")
                return []
                
    except aiohttp.ClientError as e:
        logger.error(f"Error fetching municipality {municipality_id}: {str(e)}")
        return []

async def get_municipality_data(end_index: int = 645, batch_size: int = 50) -> List[Dict]:
    end_index += 1
    municipality_data = []
    
    conn = aiohttp.TCPConnector(limit=batch_size)
    timeout = aiohttp.ClientTimeout(total=30)

    async with aiohttp.ClientSession(connector=conn, timeout=timeout) as session:
        for batch_start in range(1, end_index, batch_size):
            batch_end = min(batch_start + batch_size, end_index)
            
            tasks = [
                fetch_municipality(session, municipality_id) 
                for municipality_id in range(batch_start, batch_end)
            ]
            
            logger.info(f"Fetching municipalities {batch_start} to {batch_end-1}")
            
            results = await asyncio.gather(*tasks, return_exceptions=True)
            
            for result in results:
                if isinstance(result, Exception):
                    logger.error(f"Task error: {str(result)}")
                    continue
                municipality_data += [result]
                
            logger.info(f"Completed batch. Total records so far: {len(municipality_data)}")
    
    return municipality_data

def calculate_crime_rate_change(data, field):
    sorted_data = sorted(data, key=lambda x: x['ano'], reverse=True)
    
    changes = []
    for i in range(len(sorted_data)-1):
        current_year = sorted_data[i]
        prev_year = sorted_data[i+1]
        
        current_total = current_year[field]
        prev_total = prev_year[field]
        
        if prev_total > 0:  # Avoid division by zero
            pct_change = (current_total - prev_total) / prev_total
        else:
            pct_change = float('inf') if current_total > 0 else 0
            
        changes.append({
            'year': current_year['ano'],
            'change': pct_change
        })
    
    return changes
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

def prepare_data(cities_with_lpr, cities_without_lpr, city_names):
    """
    Prepare the data for analysis by combining both datasets
    """
    data_rows = []
    
    # Process cities with LPR
    for city_idx, city_data in enumerate(cities_with_lpr):
        for year_data in city_data:
            data_rows.append({
                'city': city_names[city_idx],
                'year': year_data['ano'],
                'vehicle_crimes': year_data['furtoVeiculo'] + year_data['rouboVeiculo'] + year_data['frv'],
                'total_crimes': year_data['furto'] + year_data['roubo'] + year_data['homicidio'],
                'has_lpr': True,
                'crime_rate': (year_data['furtoVeiculo'] + year_data['rouboVeiculo'] + year_data['frv']) / 
                             (year_data['furto'] + year_data['roubo'] + year_data['homicidio'] + 1)  # Adding 1 to avoid division by zero
            })
    
    # Process cities without LPR
    for city_idx, city_data in enumerate(cities_without_lpr):
        for year_data in city_data:
            data_rows.append({
                'city': city_names[city_idx + len(cities_with_lpr)],
                'year': year_data['ano'],
                'vehicle_crimes': year_data['furtoVeiculo'] + year_data['rouboVeiculo'] + year_data['frv'],
                'total_crimes': year_data['furto'] + year_data['roubo'] + year_data['homicidio'],
                'has_lpr': False,
                'crime_rate': (year_data['furtoVeiculo'] + year_data['rouboVeiculo'] + year_data['frv']) / 
                             (year_data['furto'] + year_data['roubo'] + year_data['homicidio'] + 1)
            })
    
    return pd.DataFrame(data_rows)

def analyze_trends(df):
    """
    Analyze trends and patterns in the data
    """
    results = {}
    
    # Overall statistics
    results['summary'] = {
        'lpr_cities': {
            'avg_vehicle_crimes': df[df['has_lpr']]['vehicle_crimes'].mean(),
            'std_vehicle_crimes': df[df['has_lpr']]['vehicle_crimes'].std(),
            'avg_crime_rate': df[df['has_lpr']]['crime_rate'].mean()
        },
        'non_lpr_cities': {
            'avg_vehicle_crimes': df[~df['has_lpr']]['vehicle_crimes'].mean(),
            'std_vehicle_crimes': df[~df['has_lpr']]['vehicle_crimes'].std(),
            'avg_crime_rate': df[~df['has_lpr']]['crime_rate'].mean()
        }
    }
    
    # Year-by-year comparison
    yearly_stats = df.groupby(['year', 'has_lpr']).agg({
        'vehicle_crimes': ['mean', 'std'],
        'crime_rate': 'mean'
    }).round(2)
    
    results['yearly_trends'] = yearly_stats
    
    # Statistical tests
    # Mann-Whitney U test (non-parametric comparison)
    statistic, pvalue = stats.mannwhitneyu(
        df[df['has_lpr']]['vehicle_crimes'],
        df[~df['has_lpr']]['vehicle_crimes'],
        alternative='two-sided'
    )
    results['mann_whitney_test'] = {'statistic': statistic, 'pvalue': pvalue}
    
    # Calculate year-over-year changes
    for has_lpr in [True, False]:
        yearly_means = df[df['has_lpr'] == has_lpr].groupby('year')['vehicle_crimes'].mean()
        results[f'{"lpr" if has_lpr else "non_lpr"}_yoy_changes'] = yearly_means.pct_change()
    
    return results

def create_visualizations(df):
    """
    Create multiple visualizations to understand patterns
    """
    figs = []
    
    # 1. Time series plot
    fig1, ax1 = plt.subplots(figsize=(12, 6))
    sns.lineplot(data=df, x='year', y='vehicle_crimes', hue='has_lpr', marker='o')
    ax1.set_title('Vehicle Crimes Over Time: LPR vs Non-LPR Cities')
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Average Vehicle Crimes')
    figs.append(fig1)
    
    # 2. Box plot comparison
    fig2, ax2 = plt.subplots(figsize=(10, 6))
    sns.boxplot(data=df, x='year', y='vehicle_crimes', hue='has_lpr')
    ax2.set_title('Distribution of Vehicle Crimes by Year')
    ax2.set_xlabel('Year')
    ax2.set_ylabel('Number of Vehicle Crimes')
    plt.xticks(rotation=45)
    figs.append(fig2)
    
    # 3. Crime rate comparison
    fig3, ax3 = plt.subplots(figsize=(10, 6))
    sns.boxplot(data=df, x='year', y='crime_rate', hue='has_lpr')
    ax3.set_title('Vehicle Crime Rate (Relative to Total Crimes)')
    ax3.set_xlabel('Year')
    ax3.set_ylabel('Vehicle Crime Rate')
    plt.xticks(rotation=45)
    figs.append(fig3)
    
    return figs

def generate_report(results):
    """
    Generate a comprehensive report of the findings
    """
    report = f"""
    LPR Effectiveness Analysis Report
    ===============================
    
    Overall Comparison:
    
    Cities with LPR:
    - Average vehicle crimes: {results['summary']['lpr_cities']['avg_vehicle_crimes']:.2f}
    - Standard deviation: {results['summary']['lpr_cities']['std_vehicle_crimes']:.2f}
    - Average crime rate: {results['summary']['lpr_cities']['avg_crime_rate']:.3f}
    
    Cities without LPR:
    - Average vehicle crimes: {results['summary']['non_lpr_cities']['avg_vehicle_crimes']:.2f}
    - Standard deviation: {results['summary']['non_lpr_cities']['std_vehicle_crimes']:.2f}
    - Average crime rate: {results['summary']['non_lpr_cities']['avg_crime_rate']:.3f}
    
    Statistical Significance:
    - Mann-Whitney U test p-value: {results['mann_whitney_test']['pvalue']:.4f}
    
    Key Findings:
    1. {'There is' if results['mann_whitney_test']['pvalue'] < 0.05 else 'There is not'} a statistically significant difference between LPR and non-LPR cities
    2. The relative crime rate (vehicle crimes / total crimes) is {'higher' if results['summary']['lpr_cities']['avg_crime_rate'] > results['summary']['non_lpr_cities']['avg_crime_rate'] else 'lower'} in cities with LPR
    """
    return report

def detect_change_points(df):
    """
    Attempt to detect significant changes in trends that might indicate LPR implementation
    """
    from scipy.stats import ttest_ind
    
    changes = []
    years = sorted(df['year'].unique())
    
    for year in years[1:]:  # Skip first year
        before = df[df['year'] < year]
        after = df[df['year'] >= year]
        
        # Compare means before and after for LPR cities
        lpr_before = before[before['has_lpr']]['vehicle_crimes'].mean()
        lpr_after = after[after['has_lpr']]['vehicle_crimes'].mean()
        
        # Calculate relative change
        if lpr_before > 0:
            pct_change = (lpr_after - lpr_before) / lpr_before * 100
            changes.append({
                'year': year,
                'percent_change': pct_change,
                'before_mean': lpr_before,
                'after_mean': lpr_after
            })
    
    return pd.DataFrame(changes)

def run_complete_analysis(cities_with_lpr, cities_without_lpr, city_names):
    # 1. Prepare the data
    print("Preparing data...")
    df = prepare_data(cities_with_lpr, cities_without_lpr, city_names)
    
    # 2. Run the trend analysis
    print("\nAnalyzing trends...")
    results = analyze_trends(df)
    
    # 3. Create visualizations
    print("\nCreating visualizations...")
    figures = create_visualizations(df)
    
    # 4. Generate report
    print("\nGenerating report...")
    report = generate_report(results)
    
    # 5. Look for potential change points
    print("\nAnalyzing potential implementation periods...")
    change_points = detect_change_points(df)
    
    return {
        'dataframe': df,
        'results': results,
        'figures': figures,
        'report': report,
        'change_points': change_points
    }

analysis_results = run_complete_analysis(top_sentry_municipalities_data, non_sentry_municipalities_data, city_names)
```
